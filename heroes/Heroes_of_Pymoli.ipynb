```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
# display total number of players
players=purchase_data["SN"].nunique()

purchase_data["Total Players"]=players
purchase_data.iloc[[0],[7]].head()
```

```python
#number of unique items
items=purchase_data["Item ID"].nunique()
purchase_data["Number of Unique Items"]=items
avgprice=purchase_data["Price"].mean()
rndprice=("$""%.2f"%(avgprice))
purchase_data["Average Price"]=rndprice
purch=purchase_data["Item ID"].count()
purchase_data["Number of Purchases"]=purch
rev=(int(purch)*int(avgprice))
rev1="$"+str(rev)
purchase_data["Total Revenue"]=str(rev1)

purchase_data.iloc[[0],[8,9,10,11]].head()
```

```python
gender=purchase_data[["SN","Gender"]]
gender=gender.drop_duplicates()
gender_count = gender["Gender"].value_counts()

# List of values
total_counts = [gender_count[0],gender_count[1],gender_count[2]]
percents_g = [round((gender_count[0]/players)*100,2),round((gender_count[1]/players)*100,2),round((gender_count[2]/players)*100,2)]

# Creating DataFrame & setting index
gender_df = pd.DataFrame({
    "Percentage of Players": percents_g,
    "Total Count": total_counts
})
gender_df.index = (["Male", "Female", "Other / Non-Disclosed"])
gender_df
```

```python
gdf=purchase_data.groupby("Gender")
sumitems=gdf["Price"].sum()
gitems=gdf["Price"].count()
avgtotal=gdf["Price"].mean()
gender_df["Purchase Count"]=total_counts
gender_df["Average Purchase Price"]=round((sumitems/gitems),2)
gender_df["Total Purchase Price"]=(sumitems)
gender_df["AVG Total Purchase per Person"]=sumitems/gender_df["Total Count"]
gender_df.loc[["Male", "Female", "Other / Non-Disclosed"],["Purchase Count","Average Purchase Price","Total Purchase Price","AVG Total Purchase per Person"]]
```

```python
bins=[0,10,15,20,25,30,35,40,500]
ranges=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Ranges"]=pd.cut(purchase_data["Age"],bins, labels=ranges)
age_df=purchase_data.groupby("Age Ranges")
age_count=age_df.count()

age_percent=age_count["Total Players"]/age_count["Total Players"].sum()



age_count["Age Percentage"]=(age_percent*100).round(2)
age_count.loc[["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"],["Age", "Age Percentage"]]
```

```python
age_pcount=age_df["Price"].count()
avg_age_price=age_df["Price"].mean()
tot_age_val=age_df["Price"].sum()
total_age_count=purchase_data["Age Ranges"].value_counts()

age_df2 = pd.DataFrame({})
age_df2["Purchase Count"]=age_pcount
age_df2["Average Purchase Price"]=avg_age_price
age_df2["Total Purchase Value"]=tot_age_val
age_df2["AVG Total Purchase per Person"]=tot_age_val/total_age_count
age_df2.head()
#age_df.head()
```

```python
#top spenders
purch_count=purchase_data.groupby(["SN"]).count()["Price"]
avg_purch_price=purchase_data.groupby(["SN"]).mean()["Price"]
tot_purch_val=purchase_data.groupby(["SN"]).sum()["Price"]

spenders=pd.DataFrame({"Purchase Count": purch_count,"Average Purchase Price":avg_purch_price,"Total Purchase Value":tot_purch_val})

spenders.sort_values("Purchase Count", ascending= False).head(5)
```

```python
#most popular items
items_purch_count=purchase_data.groupby(["Item ID","Item Name"]).count()["Price"]
items_name=purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"]
items_purch_val=purchase_data.groupby(["Item ID","Item Name"]).sum()["Price"]
pop_items=pd.DataFrame({"Purchase Count":items_purch_count,"Item Price":items_name,"Total Purchase Value":items_purch_val})
pop_items.sort_values("Purchase Count",ascending=False).head(5)
```

```python
#most profitable items
items_purch_count=purchase_data.groupby(["Item ID","Item Name"]).count()["Price"]
items_name=purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"]
items_purch_val=purchase_data.groupby(["Item ID","Item Name"]).sum()["Price"]
pop_items=pd.DataFrame({"Purchase Count":items_purch_count,"Item Price":items_name,"Total Purchase Value":items_purch_val})
pop_items.sort_values("Total Purchase Value",ascending=False).head(5)
```
